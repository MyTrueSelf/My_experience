```python
import os
import re
import csv
os.chdir(r"C:\Users\74857\Desktop\과제\삼성물산")
os.getcwd()
hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
word_counts=dict() #카운팅을 위한 딕셔너리 
with open(r"소재.csv",'r',newline='',encoding="UTF-8") as csv_input_file:
    filereader=csv.reader(csv_input_file)
    header=next(filereader)
    print(header[1])
    for row in filereader:
        if(len(row)<1):
            continue
        temp=",".join(row) #list -> string
        temp=re.split('[][+ ()%:;,._#-@】*？&$~®■|=^<>{}皮”`○™社‘㈜½　聚氨酯人造革]+',temp) #각 line에서 특수문자 제거
        for temp_list in temp:
            result=hangul.sub('',str(temp_list)) #한글 제외하고 제거
            if result:
                word_counts[result]= word_counts.get(result,0) + 1  #한글 카운팅
        
    for word,count in word_counts.items(): #카운팅 된 한글 출력
        print(word,count)
```
