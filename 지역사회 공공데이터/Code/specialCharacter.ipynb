```python
#특수문자 골라내기 위한 코드 
import re
import csv
import os
os.chdir(r"C:\Users\duswn\Desktop\빅데이터팜\삼성물산 프로젝트\06.26 진행사항")
os.getcwd()
sC = re.compile('[ ㄱ-ㅣ가-힣a-zA-Z0-9,]+') #특수문자만을 검출하기 위한 정규표현식
word_counts=dict() #카운팅을 위해 딕셔너리 생성.
with open(r"소재.csv",'r',newline='',encoding="UTF-8") as csv_input_file:
    filereader=csv.reader(csv_input_file)
    header=next(filereader)
    print(header[1])
    for row in filereader: 
        if(len(row)<1): #예외 처리
            continue
        temp=",".join(row) #join을 통해서 리스트를 문자열로 바꿔준다.
        temp=re.split('[ ㄱ-ㅣ가-힣a-zA-Z0-9,]',temp) #특수문자를 제외한 부분을 잘라낸다.
        for temp_list in temp:
            result=sC.sub('',temp_list) #특수문자를 제외한 부분을 ''로 치환.
            if result: #특수문자가 존재하면
                word_counts[result]=word_counts.get(result,0) + 1 #딕셔너리에 추가.
        
    for word,count in word_counts.items():
        print(word,count) #딕셔너리 출력 
```
