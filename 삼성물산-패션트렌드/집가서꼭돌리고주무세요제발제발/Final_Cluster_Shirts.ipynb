```python
from sklearn import preprocessing
import numpy as np
import pandas as pd

le = { key: preprocessing.LabelEncoder() for key in rawData.columns } 

for key, encoder in le.items():
    encoder.fit(list(rawData[key].values))
    rawData[key] = encoder.transform(rawData[key])
```

```python
# TODO: create a OneHotEncoder object, and fit it to all of X

# 1. INSTANTIATE
enc = preprocessing.OneHotEncoder()

# 2. FIT
enc.fit(rawData)
```

```python
# 3. Transform
onehotlabels = enc.transform(rawData).toarray()
onehotlabels.shape

# as you can see, you've the same number of rows 891
# but now you've so many more columns due to how we changed all the categorical data into numerical data
```

```python
onehotlabels
```

```python
df = pd.DataFrame(onehotlabels)
```

```python
df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt

ks = range(1,11)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df)
    inertias.append(model.inertia_)
    del model
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
